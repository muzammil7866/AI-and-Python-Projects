```python
units = int(input("Enter the number of units of your electricity bill: "))

price = 0

if(units <= 50):
    price = units*.5
    price += 20/100*price
    

elif(units > 50 and units <= 150):
    price = 50*.5 + (units-50)*.75
    price += 20/100*price
    
elif(units > 150 and units <= 250):
    price = 50*.5 + 100*.75 + (units-150)*1.2
    price += 20/100*price
    
else:
    price = 50*.5 + 100*.75 + 100*1.2 + (units-250)*1.5
    price += 20/100*price
    
print(price)
```

```python
number = int(input("Enter the number to check if it is a perfect number or not: "))
sum = 0
number2 = (number//2)+1

for i in range(1, number2):
    if(number%i == 0):
        sum += i
    
if(sum == number):
    print("Your number is a perfect number!")
    
else:
    print("Your number is not a perfect number!")
```
