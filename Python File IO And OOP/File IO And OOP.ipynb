```python
#QUESTION 1

def displayFile(fHand):
    for i in fHand:
        print(i.rstrip())
        
def countWords(fHand):
    count = 0
    for i in fHand:
        line = i
        if(line[0]!= '\n'):
            for i in line:
                if i == ' ' or i == '\n':
                    count += 1
    return count+1     

def countLines(fHand):
    count = 0
    for i in fHand:
        count += 1
        
    return count

def appendToFile(fHand):
    text = input("ENTER THE CONTENTS THAT YOU WANT TO APPEND TO THE FILE: ")
    fHand.write(text)
    
try:
    fHand = open("F223110.txt", 'r')
    displayFile(fHand)
    fHand.close()
    
    fHand = open("F223110.txt", 'r')
    print(countWords(fHand))
    fHand.close()
    
    fHand = open("F223110.txt" , 'r')
    print(countLines(fHand))
    fHand.close()
    
    fHand = open("F223110.txt", 'a')
    appendToFile(fHand)
    fHand.close()
    
    fHand = open("F223110.txt", 'r')
    displayFile(fHand)
    fHand.close()
    
    
    
    
except FileNotFoundError:
    print("FILE DOES NOT EXIST!")
    
    
    
finally:
    print("DONE!")
```

```python
#QUESTION 2

class Bank:
    def __init__(self, Accounts):
        self.Accounts = Accounts
        
    def __str__(self):
        str1 = ""
        for i,j in self.Accounts.items():
            str1 = str1 + (f"{i} {j} \n")
            
        return str1
            
            
class Account:
    def __init__(self, holderName, initialBalance):
        self.holderName = holderName
        self.balance = initialBalance
        
    def deposit(self, amount):
        self.calculateCurrentBalance(1,0,amount)
    
    def withdraw(self, amount):
        self.calculateCurrentBalance(0,1,amount)
    
    def calculateCurrentBalance(self, d, w, amount):
        if(w == 1):
            self.balance -= amount
            
        elif(d == 1):
            self.balance += amount
    
    def __str__(self):
        return (f"{self.holderName} {self.balance}") 
        

class Transaction:
    def __init__(self, accNo, transAmt):
        self.accNo = accNo
        self.transAmt = transAmt
        
    
    def executeTransaction(self, Accounts, transactionDetails):
        
        option = int(input("ENTER 1 FOR DEPOSIT, ENTER 2 FOR WITHDRAW: "))
        
        found = False
        
        for i,j in Accounts.items():
            if(i == self.accNo):
                found = True
                
        if(found == True):
            if(option == 1):
                Accounts[self.accNo].deposit(self.transAmt)
                transactionDetails.append(f"DEPOSITED {self.transAmt} INTO ACC NO.:{self.accNo}")
                
            elif(option == 2):
                Accounts[self.accNo].withdraw(self.transAmt)
                transactionDetails.append(f"WITHDRAWN {self.transAmt} FROM ACC NO.:{self.accNo}")
                
        else:
            print("ACCOUNT NOT FOUND! KINDLY CREATE AN ACCOUNT FIRST TO PERFORM TRANSACTION.")

    def __str__(self):
        return (f"{self.accNo} {self.transAmt}")

        
        
transactionDetails = []
Accounts = {}
bank1 = Bank(Accounts)

accNoStart = 0
end = False
option = 0

while(end!= True):
    option = int(input("ENTER 1 FOR CREATING AN ACCOUNT, ENTER 2 TO PERFORM TRANSACTION, ENTER 3 FOR VIEWING TRANSACTION HISTORY, ENTER 4 FOR EXITTING THE PROGRAM: "))
    
    if(option == 1):
        acc2 = Account("", 0)
        acc2.holderName = input("ENTER THE ACCOUNT HOLDERS NAME: ")
        acc2.balance = int(input("ENTER THE INITAL DEPOSIT AMOUNT: "))
        acc2 = Account(acc2.holderName, acc2.balance)
        accNoStart += 1
        bank1.Accounts.setdefault(accNoStart, acc2)
        
        print(acc2)
        print(bank1)
        
    elif(option == 2):
        trans2 = Transaction(0, 0)
        trans2.accNo = int(input("ENTER THE INTENDED ACCOUNT NUMBER: "))
        trans2.transAmt = int(input("ENTER THE TRANSACTION AMOUNT: "))
        trans2 = Transaction(trans2.accNo, trans2.transAmt)
        trans2.executeTransaction(Accounts, transactionDetails)
            
        print(trans2)
        print(bank1)
        
        
    elif(option == 3):
        print("TRANSACTION HISTORY: ")
        
        for i in transactionDetails:
            print(i)
            
        print(bank1)
            
    elif(option == 4):
        end = True
        
    else:
        print("INVALID INPUT!")
        
        
```

```python
#QUESTION 5
def square(l1):
    l1 = list(map(lambda x: x**2, l1))
    print(l1)

def vowels(l2):
    l2 = list(filter(lambda x: x[0] == 'a' or x[0] == 'e' or x[0] == 'i' or x[0] == 'o' or x[0] == 'u', l2)) 
    print(l2)
    
def reverseStr(text):
    size = len(text)
    size = size-1
    text2 = ''
    for i in range(size, -1, -1):
        text2 = text2 + text[i]
     
    print(text2)
    return text2
    
square([1,2,3,4])
vowels(['he', 'egg', 'mr.', 'after', ])
reverseStr("hello")
```
